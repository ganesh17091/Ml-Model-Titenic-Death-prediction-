```python
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

from sklearn.preprocessing import StandardScaler




class Dropfeture(BaseEstimator, TransformerMixin):

    def fit(self,X,y=None):
        return self

    def transform(self,X):
        X = X.drop(['Name', 'Ticket','Cabin'], axis = 1)
        return X




class AgeImputer(BaseEstimator, TransformerMixin):
    
    def fit(self,X,y=None):
        return self

    
    def transform(self,X):
        impute = SimpleImputer(strategy = 'mean')
        X['Age'] = impute.fit_transform(X[['Age']])
        return X
        

class encoder(BaseEstimator, TransformerMixin):

    def fit(self,X,y = None):
        return self

    def transform(self, X):
        
        encode = OneHotEncoder(sparse_output = False)
        
        X1 = encode.fit_transform( pd.DataFrame(X[['Sex','Embarked']]) )

        X = X.drop(['Sex','Embarked'], axis = 1)
        X.columns = X.columns.astype(str)

        return pd.concat( [X, pd.DataFrame(X1)], axis = 1)



class Scalling(BaseEstimator, TransformerMixin):

    def fit(self, X, y = None):
        return self

    def transform(self, X):
        stdscalar = StandardScaler()
        X = stdscalar.fit_transform(X)
        X.columns = X.columns.astype(str)
        return X 




        
                

    
```

```python
from sklearn.pipeline import Pipeline
preprocess = Pipeline([

('dropper',Dropfeture()),
('encoder',encoder()),
('Age_imputer',AgeImputer())

])


Scalling_pipe = Pipeline([

('Scalling', Scalling())

])
```
